### Missing Value Ratio

```python
#lets keep threeshold value as 0.7

for i in data.columns:
    tmp=data[i].isnull().sum()/len(data)
    if tmp >0.7:
        print("Removing Column : "+i +"with missing value ratio",tmp)
        data=data.drop(columns=[i])
```
