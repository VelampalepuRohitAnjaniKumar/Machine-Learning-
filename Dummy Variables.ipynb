```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Transformed_Dataset.csv")
df.head()
```

```python
df=df.rename(columns={'Sale Price':'Sale_Price','Condition of the House':'Condition_of_house','Ever Renovated':'Ever_renovated','Waterfront View':'Waterfront_view'})

df.head()
```

# Dummies

```python
#converting categorical into numerical variables

df=pd.get_dummies(df,columns=['Condition_of_house'],drop_first=True)
df
```

```python
df=pd.get_dummies(df,columns=['Ever_renovated','Waterfront_view'],drop_first=True)

df.head()
```

# Binning


```python
zip_table=df.groupby('Zipcode').agg({'Sale_Price':"mean"}).sort_values('Sale_Price',ascending=True)
zip_table.head()
```

```python
zip_table['Zipcode_Group']= pd.cut(zip_table['Sale_Price'], bins = 10,
                                    labels = ['Zipcode_Group_0',
                                              'Zipcode_Group_1',
                                              'Zipcode_Group_2',
                                              'Zipcode_Group_3',
                                              'Zipcode_Group_4',
                                              'Zipcode_Group_5',
                                              'Zipcode_Group_6',
                                              'Zipcode_Group_7', 
                                              'Zipcode_Group_8',
                                              'Zipcode_Group_9'
                                             ],
                                    include_lowest = True)

```

```python
zip_table.head()
```

```python
zip_table=zip_table.drop(columns='Sale_Price')
```

# Merging Tables

```python
df=pd.merge(df,zip_table,
            left_on='Zipcode',
            how='left',
            right_index=True)
df.head()
```

```python
df=df.drop(columns='Zipcode_Group_y')
df.head()
```

## Dummies for Zipcode_group_x

```python
df=pd.get_dummies(df,columns=['Zipcode_Group_x'],drop_first=True)
df.head()
```

```python
df.to_csv('Tranformed_Data_2.csv')
```
